#### Q4) Find the min max for beats.per.minute and liveness
varance and standard deviation fro energy and dancibility
based on top 10 genres

```python
top_genres = df['top genre'].value_counts()
top_genres = top_genres.sort_values(ascending = False).head(10)
top_genres 
```

```python
top_genres.index
```

```python
ex4 = df[df['top genre'].isin(top_genres.index)].groupby(['top genre']).agg({'beats.per.minute':['min','max'],
                                            'liveness':['min','max'],
                                              'energy':['var','std'],  
                                              'danceability':['var','std']})
ex4
```
